# Combinar databases

```python
# Importaciones
import pandas as pd

import sys
sys.path.append('..')
from src import sp_limpieza as sp
```

```python
# Leer los archivos 
df_online = pd.read_excel('../data/Online Retail.xlsx')
df_sales = pd.read_excel('../data/Sales Records.xlsx')
```

```python
# Combinar databases
df_combined = pd.concat([df_online, df_sales], axis=1)

```

```python
# Guardar el dataframe combinado
df_combined.to_excel('../data/datos.xlsx', index=False)
```
