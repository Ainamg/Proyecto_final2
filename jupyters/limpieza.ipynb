```python
import pandas as pd

import sys
sys.path.append('..')
from src import sp_limpieza as sp
```

```python
df = pd.read_excel('../data/datos.xlsx')
df.sample(5)
```

```python
# Cambiar el nombre de las columnas 
df.columns = df.columns.str.replace(' ', '_').str.lower()
df.sample(5) 
```

```python
# Poner todo en minúsculas
sp.valores_minus(df)
```

```python
# Quitar decimales de la columna customerid
df['customerid'] = df['customerid'].astype('Int64')
```

```python
# Convertir las columnas de fecha a formato datetime
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
df['ship_date'] = pd.to_datetime(df['ship_date'], errors='coerce')
```

```python
df['ship_year'] = df['ship_date'].dt.year
df['ship_month'] = df['ship_date'].dt.month
df['ship_day'] = df['ship_date'].dt.day_name()
df['ship_quarter'] = df['ship_date'].dt.quarter
```

```python
df.to_excel('../data/datos_limpios.xlsx', index=False)
```
